```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1  
```

```python
list_of_simple_preprocess_data = []
for i in list_of_sentence:
    list_of_simple_preprocess_data.append(gensim.utils.simple_preprocess(i, deacc=True, min_len=3))
```

```python
bigram = gensim.models.Phrases(list_of_simple_preprocess_data) 
```

```python
bigram
```

```python
from gensim.utils import lemmatize
from nltk.corpus import stopwords
```

```python
stops = set(stopwords.words('english')) 
```

```python
from nltk.stem import WordNetLemmatizer
wordnetlemmatizer = WordNetLemmatizer()
```

```python
def process_texts(texts):
    texts = [[word for word in line if word not in stops] for line in texts]
    texts = [bigram[line] for line in texts]
    return texts
```

```python
import re
tr_texts = process_texts(list_of_simple_preprocess_data)

```

```python
train_texts = []
for i in range(len(tr_texts)):
  for word in tr_texts[i]:
    train_texts.append(wordnetlemmatizer.lemmatize(word))
```

```python
from gensim.models import LdaModel
from gensim.models.wrappers import LdaMallet
from gensim.corpora import Dictionary
```

```python
train_texts
```

```python
train_texts = [d.split() for d in train_texts]
dictionary = Dictionary(train_texts)
corpus = [dictionary.doc2bow(text) for text in train_texts]
```

```python
print(dictionary)
```

```python
print(corpus)
```

```python
ldamodel = LdaModel(corpus=corpus, num_topics=6, id2word=dictionary)
```
